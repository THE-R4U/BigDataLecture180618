## 함수의 인자로 함수가 들어가는 경우 (매우 중요!)

```python
# Closer 기법
def get_speak_func(volume):
    def whisper(text):
        return text.lower() + '...'
    def yell(text):
        return text.upper()+'!..'
    if volume > 0.5:
        return yell
    else:
        return whisper
```

```python
get_speak_func(0.3)
```

```python
def make_adder(n):
    def add(x):
        return x + n
    return add
```

```python
make_adder(3)(6)
```
